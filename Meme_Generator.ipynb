```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import os
from google.colab import files

# Upload image file
uploaded = files.upload()
for file_name in uploaded.keys():
    image = Image.open(file_name)
    print(f"✅ Uploaded: {file_name}")
    break

def generate_meme(image, top_text, bottom_text):
    # Convert to RGB
    image = image.convert("RGB")
    width, height = image.size

    # Load font with a smaller size (Reduced to height/15)
    try:
        font_size = int(height / 25)  # Smaller font
        font = ImageFont.truetype("impact.ttf", font_size)
    except OSError:
        print("⚠️ Font not found! Please upload 'impact.ttf'.")
        return

    # Draw text
    draw = ImageDraw.Draw(image)

    # Function to add text
    def draw_text(text, y_pos):
        bbox = draw.textbbox((0, 0), text, font=font)  # Fixed: textbbox() replaces textsize()
        text_width = bbox[2] - bbox[0]  # width
        text_height = bbox[3] - bbox[1]  # height
        x = (width - text_width) / 2
        draw.text((x, y_pos), text, font=font, fill="blue", stroke_fill="pink", stroke_width=3)

    # Add top and bottom text with padding
    draw_text(top_text.upper(), int(height * 0.05))  # 5% from top
    draw_text(bottom_text.upper(), height - font_size - int(height * 0.05))  # 5% from bottom

    # Save and show meme
    image.save("meme.png")
    plt.imshow(image)
    plt.axis("off")
    plt.show()

    print("✅ Meme saved as 'meme.png'")

# Input text
top_text = input("Enter TOP text: ")
bottom_text = input("Enter BOTTOM text: ")

# Generate meme
generate_meme(image, top_text, bottom_text)
```
